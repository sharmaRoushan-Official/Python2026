# 15. Writing Files in Python
 * __To write a text file in Python, you follow these steps:__
 * __1.First, Open the text file for writing(or appending) using the open function.__
 * __2.Second, write to the text file using write(), writeline(), or writelines() method.__
 * __3.Third, close the file using the close() method.__


* __Some Characters and thier functions__

| Mode | Descriptions |
|----------|----------|
| "r"    | Open file for reading only, Starts reading from the beginning of the file. This is defualt mode.   | 
| "rb"   | Open file for reading only in binary format. Starts reading from the beginning of the text file.   | 
| "r+"    | Open file for reading and writing. File pointer placed at beginnning of the file.   | 
| "w"    | Open file for writing only. File Pointer placed at the beginning of the file.Overwrites an exisiting file and creates a new one if it does not exits.   | 
| "wb"    | Same as w but opens in binary mode.   | 
| "wb+"    | Same as wb but also allows reading text from the file.   | 
| "a"    | Open file for appending. Starts writing at the end of the file. Creates a new file if the file does not exits.   | 
| "ab"    | Same as a but in binary format. Creates a new file if file does not exits.   | 
| "ab+"    | Same4 as ab but also open for reading.   |

```python
fileName = input("Enter your file name here: ")

completeFileName = "C:\\Users\\User\\Desktop\\"+fileName+".txt"

with open(completeFileName,'w') as f:
    endPoint = int(input("Enter number of lines value: "))

    for i in range(1,endPoint+1):
        sentence = input("Enter your text here: ")
        f.write(str(i))
        f.write('.')
        f.write(sentence)
        f.write('\n')
    print("Data Saved Successfully",completeFileName)
```
