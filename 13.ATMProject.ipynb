```python
class atm:
    def __init__(self,balance=0):
        while True:
            self.name=input("Enter your name here: ")  # Omkar Pingale1
            if self.name.replace(' ',"").isalpha(): # OmkarPingale1
                break
            else:
                print("Enter Your Correct name without Digits or Numbers")
        while True:
            self.accountNum = input("Set your 8-Digits Account Number:")

            if self.accountNum.isnumeric() and len(self.accountNum) == 8:
                break
            else:
                print("Enter Correct Account Number With Length of 8 Digits.")

        while True:
            
            self._pin = input("Set Your 4-Digits PIN: ")

            if self._pin.isnumeric() and len(self._pin) == 4:
                break
            else:
                print("Enter Correct PIN Number with Length of 4-Digits.")
            
        self.balance=balance

    # this method display Account Holder Details
    def details(self):
        '''Display Details of Account Holder'''
        # 12345678
        first2Digit = self.accountNum[0:3]  # 12
        last2Digit = self.accountNum[6:]
        completeAccountNum = first2Digit+"XXXX"+last2Digit
        
        print(f"Account Holder Name:{self.name}")
        print(f"Account Number:{completeAccountNum}")
        print(f"Current Balance is:{self.balance}")

    # this method display Account Current balance.
    def showBalance(self):
        pin2 = input('Enter Your Current PIN: ')

        if self._pin == pin2:
            print(f"Your Current Balance is: {self.balance}")
        else:
            print("Incorrect PIN.")

    # this method deposite cash
    def depositeCash(self):
        pin2 = input("Enter your Current PIN: ")
        if pin2 == self._pin:
            depo = int(input("Enter your Amount ₹: "))
            self.balance = self.balance + depo
            print(f"Your Updated Balance is ₹:{self.balance}")
        else:
            print("Incorrect PIN.\nEnter Correct PIN.")

    # this mthod withdraw cash 
    def wittCash(self):
        pin2 = input("Enter your Current PIN: ")

        if pin2 == self.pin:
            amount = int(input("Enter your Amount to withdraw ₹:"))
    
            if amount<self.balance:
                self.balance =self.balance - amount
                print(f"Collect your cash ₹: {amount}\nYour Updated Balance is ₹:{self.balance}")
            else:
                print(f"Insufficient Balance ₹:{self.balance}")
        else:
            print('Invalid PIN.\nEnter Correct PIN')


    # this method to change Name
    def changeName(self):
        oldName = input("Enter your old Name:")
        if oldName == self.name:
            newName = input("Enter your New Name:")
            if newName != oldName:
                if newName.replace(" ",'').isalpha():
                    print(newName)
                    self.name = newName
                else:
                    print("New Name must be Only Alphabets Not Digits or Numbers.")
            else:
                print("Account Hold New Name and Old Name can't be same.")
        else:
            print("Invalid Current User Name.")
        
    def trans(self):
        print("""
        \n1.Account Details
        \n2.Account Balance
        \n3.Deposite Cash
        \n4.Withdraw Cash
        \n5.Change Name
        \n6.ShowAccount Number
        \n7.Change PIN
        \n8. Exit
        """)
        while True:
            try:
                opt = int(input("Choose an options: "))
            except Exception as e:
                print(f"Error: {e}")
                continue
            else:
                if opt == 1:
                    atm.details(self)
                elif opt == 2:
                    atm.showBalance(self)
                elif opt == 3:
                    atm.depositeCash(self)
                elif opt == 4:
                    atm.wittCash(self)
                elif opt == 5:
                    atm.changeName(self)

                elif opt == 8:
                    print("Thank you for Transaction!!")
                    break
obj = atm()
obj.trans()
```
