# 17. List Comprehensions 

 * __List comprehensions in python is an easy and compact syntax for creating a list from a string or another list.__
 * __It is very concise way to create a new list by operating on each item in the existing list.__
 * __List comprehension is considerably faster than processing a list using the for loop.__

```python
# x = 3  10

(lambda x:x+3)(10)
```

```python
list1 = []
for i in range(1,11):
    sqr = i**2
    list1.append(sqr)
    # print(sqr)
print(list1)
```

```python
# list comprehension

[i**2 for i in range(1,11)]
```

```python
list2 = []
for i in range(1,11):
    if i%2==0:
        # print(i)
        list2.append(i)
print(list2)
```

```python
[i for i in range(1,11) if i%2==0]
```

```python
languages = ['Python',"C++","C","Java","Ruby on Rails","ForTran","BASIC","Kotlin","R","Go","C#"]
# C values 
languages = ['Python',"C++","C", "Java", "Ruby on Rails", "ForTran", "BASIC", "Kotlin", "R", "Go","c#"]

[i for i in languages if i[0]=='c' or i[0]=='C' ]
```
