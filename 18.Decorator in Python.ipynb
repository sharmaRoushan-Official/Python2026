# 18. Decorator in Python

 * __Decorator provide a simple, syntax for calling higher-order functions.__
 * __By definition, a decorator is a funciton that takes other functions and extends their behaviours without modification them.__
 * __Decorator can be extremely useful as they allows the extension of an existing functions, without any modification to the original function source code.__
 * __In fact, there are two types of decorator in Python:__
   * __a).ClassDecorator__
   * __b).Functiondecorator__
 * __In application, decorator are mainly used in creating a middle layer in the backend, which perform token authentication,image compression, validation, and manymore.__

```python
def sumValue(x=20):
    num1 = int(input("Enter first Number: "))
    sum1 = num1 + x
    print('Sum of the value is:',sum1)
sumValue(20)
```

```python
def multNum(y):
    num1 = int(input("Enter first Number: "))
    sum1 = num1 * y
    print('Multiplication of the value is:',sum1)

sumValue(20)
```

```python
def operation(func,x1):  # sumValue, 8
    num = func(x1)  # sumValue(8)
    return num

operation(sumValue,8)
```

```python
def divNum(x):
    y = int(input("Enter second value: "))
    result = x/y
    print("Result is:",result)
    
```

```python
def operator(func):
    try:
        a = int(input("Enter first value: "))
        val = func(a)
    except Exception as e:
        print(e)
    else:
        return val
operator(divNum)
```

```python
def addition():
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    sum1 = num1+num2
    return sum1
    
def calledFunction(func):
    try:
        res = func() # addition()
        b = int(input("Enter number for division value: "))
        div = res/b
    except Exception as e:
        print(e)
    else:
        print('your result is:',div)
calledFunction(addition)
```

```python
def decoratedTex(val): # text1, text2
    #print(val)  # display
    def text():
        function = val() #text1() , text2()
        textUpper = function.upper()
        return f"Your text is in upper case: {textUpper}"
    return text
```

```python
@decoratedTex
def text1():
    name = input("Enter string value: ")
    return name
text1()
```

```python
@decoratedTex
def text2():
    value = "amita"
    return value

text2()
```
