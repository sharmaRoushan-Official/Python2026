# 14. Reading Files in python

 * __To read a text file in python, you follow these steps:__
 * __First, Open a text file for reading by using the open() function.__
 * __Second, read text from the file using read(), realdine(), readlines() method of the file object.__
 * __Third, close the file using the file close() method, which fress up resources and ensures consistency across different python version.__

 |Method|Descriptions|
 |-----|-----|
 |__writable():__|Returns wheather the file can be written or not.|
 |__readable():__|returns whether the file stream can be read or not.|
 |__read():__| Returns the file content.|
 |__readline():__|Return one line from the file.|
 |__readlines():__|Returns a list of strings from the file.|
 |__write():__| writes the specified string to the file.|
 |__writelines():__| writes a list of strings to the file.|
 |__close():__| close the file.|
 |__flush():__|flushed the inter buffer.|
 |__seek():__|change the file position.|
 |__tell():__|Returns the current file.|
 |__truncate():__| Resize the file specified.|

```python
fileName = "C:\\Users\\User\\Desktop\\file-sample_150kB.pdf"
print(fileName)

with open(fileName,'rb') as file:
    reader = PyPDF2.PdfReader(file)
    # print(reader)

    print(f"Number of pages: {len(reader.pages)}")

    for page_num, page in enumerate(reader.pages):
        text = page.extract_text()
        print(f"------Page {page_num+1}-------")
        print(f"{text}")
```

```python
fileName = "C:\\Users\\User\\Desktop\\file-sample_150kB.pdf"
print(fileName)

with open(fileName,'rb') as file:
    reader = PyPDF2.PdfReader(file)
    # print(reader)
    print(f"Number of pages: {len(reader.pages)}")
    charCount = 0
    spaceCount = 0
    for page_num, page in enumerate(reader.pages):
        text = page.extract_text()
        
        print(f"------Page {page_num+1}-------")
        # print(f"{len(text)}")
        for i in text:
            # print(i)
            charCount+=1
            if i == " ":
                spaceCount +=1
    print("Total characters in the pdf file is:",charCount)
    print("Total spaces is:",spaceCount)
```
